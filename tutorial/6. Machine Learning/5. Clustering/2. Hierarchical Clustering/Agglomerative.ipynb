```python
# Importing the libraries  
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
```

```python
# create a blob of 200 data points
datasets = make_blobs(n_samples=200,
                     n_features=2,
                     centers=4,
                     cluster_std=1.6,
                     random_state=50)
```

```python
point= datasets[0]
```

```python
plt.scatter(point[:,0],point[:,1])
```

```python
#Finding the optimal number of clusters using the dendrogram  
import scipy.cluster.hierarchy as shc  
dendro = shc.dendrogram(shc.linkage(point, method="ward"))  

plt.title("Dendrogrma Plot")  
plt.ylabel("Euclidean Distances")  
plt.xlabel("Customers")  
plt.show()  
```

```python
#training the hierarchical model on dataset  
from sklearn.cluster import AgglomerativeClustering  
hc= AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')  
y_pred= hc.fit_predict(point)  
```

```python
# centroids = hc.cluster_centers_
# centroids_pca = pca_2.transform(centroids)

pd.Series(y_pred).value_counts()
```

```python
plt.scatter(x=point[:, 0], y=point[:, 1], c=y_pred, cmap='autumn')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Clustered Data',fontweight='bold')
plt.show()
```
