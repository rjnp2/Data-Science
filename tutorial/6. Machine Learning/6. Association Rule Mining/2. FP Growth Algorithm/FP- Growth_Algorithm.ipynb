# Implementation of FP- Growth Algorithm

To implement this, we have a problem of a retailer, who wants to find the association between his shop's product, so that he can provide an offer of "Buy this and Get that" to his customers.

The retailer has a dataset information that contains a list of transactions made by his customer. In the dataset, each row shows the products purchased by customers or transactions made by the customer. 

### To solve this problem, we will perform the below steps:
1. Data Pre-processing
1. Training the Apriori model on the dataset
3. Visualizing the results

### 1. Data Pre-processing Step:

```python
# Importing the libraries
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
```

```python
#Importing the dataset  
dataset = pd.read_csv('store_data.csv', header=None)  
dataset.head()
```

```python
from tqdm import tqdm
records = []
for i in tqdm(range(0, 7501)):
    records.append([str(dataset.values[i,j]) for j in range(0, 20) if dataset.values[i,j] is not np.NaN  ] )   
```

```python
records[:2]
```

## 2. Training the frequent patterns Model on the dataset
To train the model, we will use the pyfpgrowth  that will be imported from the pyfpgrowth package. This function will return the rules to train the model on the dataset. Consider the below code:

import pyfpgrowth

## Generating the frequent patterns

patterns = pyfpgrowth.find_frequent_patterns(transactions, min_support) 

Given a set of transactions, find the patterns in it over the specified support threshold. \
transactions: A list of transactions.\
min_support= To set the minimum support float value. Here we have used 3 that is calculated by taking 4 transactions per customer each week to the total number of transactions.

```python
#Step 4: 
import pyfpgrowth

patterns = pyfpgrowth.find_frequent_patterns (records, 4)
```

## Generate Association Rules

patterns = pyfpgrowth.generate_association_rules(patterns, confidence_threshold)

Given a set of frequent itemsets, return a dict of association rules in the form {(left): ((right), confidence)} \

patterns: A list of patterns.\
confidence_threshold= Trules = pyfpgrowth.generate_association_rules(patterns,0.8)o set the minimum confidence value

```python
rules = pyfpgrowth.generate_association_rules(patterns,0.6)
```
